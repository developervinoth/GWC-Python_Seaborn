## Importing Seaborn

```python
import seaborn as sns
%matplotlib inline
```

## Importing Sample Data

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Dist Plot

```python
sns.distplot(tips['tip'], bins = 50, kde = True)
```

## Jointplot in Seaborn

```python
sns.jointplot('total_bill', 'tip', data = tips, kind = 'reg')
```

### Pairplot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, palette = 'coolwarm', hue='time')
```

## Categorical Plots in Seaborn

```python
tips
```

```python
sns.barplot(x = tips['sex'], y = tips['tip'], estimator = sum)
```

```python
sns.countplot(tips['day'], palette ={'Yes': '#2AFF00', 'No': '#FF4572'}, hue = tips['smoker'])
```

```python
tips[tips['sex'] == 'Female']['tip'].mean()
```

```python
sns.boxplot(x='day', y='tip', data=tips)
```

```python
tips.head(5)
```

```python
sns.violinplot(x='day', y='tip', data = tips)
```

```python
sns.stripplot(x = 'day', y = 'tip', data = tips)
```

```python
sns.swarmplot(x = 'day', y = 'tip', data =tips)
```

```python
sns.violinplot(x = 'day', y ='tip', data=tips)
sns.swarmplot(x = 'day', y = 'tip', data =tips, palette=['black'])
```
