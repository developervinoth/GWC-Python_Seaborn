## Importing Seaborn

```python
import seaborn as sns
%matplotlib inline
```

## Importing Sample Data

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Dist Plot

```python
sns.distplot(tips['tip'], bins = 50, kde = True)
```

## Jointplot in Seaborn

```python
sns.jointplot('total_bill', 'tip', data = tips, kind = 'reg')
```

### Pairplot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, palette = 'coolwarm', hue='time')
```

## Grids

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
import matplotlib.pyplot as plt
```

```python
grid  =sns.PairGrid(iris)
grid.map_diag(plt.hist)
grid.map_upper(plt.scatter)
grid.map_lower(sns.kdeplot)
```

```python
grid  =sns.PairGrid(iris)
grid.map(plt.scatter)
```

## Size & Aspect Ratio (:)

```python
plt.figure(figsize = (12,5), dpi = 600)
sns.countplot(x = 'sex', data= tips)
```

```python
sns.set_context('paper', font_scale=2)
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, height = 5, aspect = 2)
```
